```python
# libraries
import urllib.request
import re
import cv2
from PIL import Image 
import os
import pandas as pd
import youtube_dl #for youtube streaming and downloads
from glob import glob #access files inside a directory
import numpy as np
```

```python
video_lnk = []
search_keyword = input("Enter the keyword:")
html = urllib.request.urlopen("https://www.youtube.com/results?search_query="+ search_keyword+"&sp=CAMSBggFEAEYAQ%253D%253D")
video_ids = re.findall(r"watch\?v=(\S{11})", html.read().decode())
video_ids = video_ids[0:4]

for vid in video_ids:
  video_lnk.append(f"https://www.youtube.com/watch?v="+ vid)
print(video_lnk)
```

```python
for i in range(len(video_lnk)):
  url=video_lnk[i]
  print(url)
  with youtube_dl.YoutubeDL({'format':'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best'}) as ydl:
    ydl.download([url])
  print("successfully")
  i+=1
  if i>=4:
    break
```

```python
import cv2
import os

#read the video
cam=cv2.VideoCapture(r"/content/drive/MyDrive/Colab Notebooks/video1.mp4")

#giving frames name
currentframe=0
while(True):
  #reading from frame
  ret,frame=cam.read()
  if ret:
    #if video is still continuing to create images
    name='/content/drive/MyDrive/Colab Notebooks/1/frame'+str(currentframe)+'.jpg'
    print('Creating...' + name)
    #writing the extracted images
    cv2.imwrite(name,frame)
    #increasing counter to show how many frames are created
    currentframe+=1
  else:
    break

#Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```python
def create_dir(path):
  try:
    if not os.path.exists(path):
      os.makedirs(path)
  except OSError:
    print(f"ERROR: Creaitng directory with name {path}")

def save_frame(video_path, save_dir):
  name = video_path.split("/")[-1].split(".")[0]
  save_path = os.path.join(save_dir,name)
  print(name)
  create_dir(save_path)
  
  cap = cv2.VideoCapture(video_path)
  idx = 0
  while True:
    ret, frame = cap.read()
    if ret == False:
      cap.release()
      break
    cv2.imwrite(f"{save_path}/{idx}.jpg", frame)
    idx+=1
if __name__=="__main__":
  video_paths = glob(r"/content/drive/MyDrive/Colab Notebooks/*.mp4")
  save_dir = "save"

  for path in video_paths:
    save_frame(path,save_dir)
```
